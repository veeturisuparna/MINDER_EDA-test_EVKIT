```python
import pandas as pd
import neurokit2 as nk
import numpy as np
from scipy.signal import butter, filtfilt

# Your EDA signal and sampling rate
# eda_signal = np.array([...])
fs = 32  

# Frequency range to be filtered out
lowcut = 2  # Low cut frequency in Hz
highcut = 4  # High cut frequency in Hz

# Design the bandstop filter
nyq = 0.5 * fs  # Nyquist frequency
low = lowcut / nyq
high = highcut / nyq
b, a = butter(N=10, Wn=[low, high], btype='bandstop')  # Using a Butterworth filter

# Apply the filter

# eda_df = pd.read_csv('C://Users//viggi//Desktop//vig_eda_4-2-2024_rest_ice-test1.csv')
eda_df = pd.read_csv('C://Users//viggi//Desktop//vig_eda_4-2-2024_rest_harrypotter4_end.csv')
eda_df = eda_df.drop(columns=[' BIOZ_DATA[19:0]', ' BTAG[2:0]', ' LDOFF_PH', ' LDOFF_PL', ' LDOFF_NH', ' LDOFF_NL', ' BOVER', ' BUNDR'], axis='columns')
dur_sec = 6*60
start_sec = 1*60
ind_start = int(fs*start_sec)
ind_end = int(fs*dur_sec)
# eda_df_new = eda_df.iloc[ind_start:ind_end].reset_index(drop=True)
eda_df_new = eda_df.iloc[2000:4500].reset_index(drop=True)

eda_df_new.head()
```

```python
eda_df_new.columns
```

```python
# scaler = np.multiply(10^6,np.ones(len(eda_df_new[' BioZ (ohm)'].values)))
# eda_signal = np.divide(scaler,eda_df_new[' BioZ (ohm)'].values)
eda_signal = 10e6 / eda_df_new[' BioZ (ohm)'].values  # convert ohms to siemens
# eda_signal = eda_signal * 1e6  # convert siemens to microsiemens
# eda_signal = np.divide(np.ones(len(eda_df_new[' BioZ (ohm)'].values)),eda_df_new[' BioZ (ohm)'].values)
filtered_eda_signal = filtfilt(b, a, eda_signal)
eda_clean1 = nk.eda_clean(eda_signal, sampling_rate=fs, method='neurokit')
eda_clean2 = nk.eda_clean(eda_signal, sampling_rate=fs, method='biosppy')
filtered_eda_signal = nk.eda_clean(eda_signal, sampling_rate=fs, method='biosppy')

nk.signal_plot([eda_signal,filtered_eda_signal, eda_clean1, eda_clean2], labels=["Raw", "BandStop","NeuroKit", "BioSPPy"])
```
